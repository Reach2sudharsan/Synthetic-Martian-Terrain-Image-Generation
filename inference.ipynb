```python
import pandas as pd
import numpy as np

def summarize_metrics(df, label, start_epoch=None, end_epoch=None):
    """Summarize key metrics from YOLO training logs."""
    if start_epoch is not None and end_epoch is not None:
        df = df[(df['epoch'] >= start_epoch) & (df['epoch'] <= end_epoch)]

    return pd.DataFrame([{
        "Label": label,
        "Mean Train Box Loss": df["train/box_loss"].mean(),
        "Mean Train Cls Loss": df["train/cls_loss"].mean(),
        "Mean Train DFL Loss": df["train/dfl_loss"].mean(),
        "Mean Val Box Loss": df["val/box_loss"].mean(),
        "Mean Val Cls Loss": df["val/cls_loss"].mean(),
        "Mean Val DFL Loss": df["val/dfl_loss"].mean(),
        "Mean mAP50": df["metrics/mAP50(B)"].mean(),
        "Mean mAP50-95": df["metrics/mAP50-95(B)"].mean(),
        "Max mAP50": df["metrics/mAP50(B)"].max(),
        "Min mAP50": df["metrics/mAP50(B)"].min(),
    }])
```

```python
# Summary over all epochs
full_summary = pd.concat([
    summarize_metrics(syn_0_data, "syn_0"),
    summarize_metrics(syn_5_data, "syn_5"),
    summarize_metrics(syn_50_data, "syn_50")
], ignore_index=True)

# Summary over epochs 180-200
summary = pd.concat([
    summarize_metrics(syn_5_data, "syn_0 (Epochs 180-200)", 180, 200),
    summarize_metrics(syn_5_data, "syn_5 (Epochs 180-200)", 180, 200),
    summarize_metrics(syn_50_data, "syn_50 (Epochs 180â€“200)", 180, 200)
], ignore_index=True)

print("Full Epoch Summary")
print(full_summary.to_string(index=False))

print("Summary from Epochs 180â€“200")
print(summary.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
plt.figure(figsize=(12, 6))
plt.plot(syn_0_data['epoch'], syn_0_data['metrics/mAP50(B)'], label='syn_0 mAP50', color='red')
plt.plot(syn_5_data['epoch'], syn_5_data['metrics/mAP50(B)'], label='syn_5 mAP50', color='#1f77b4')
plt.plot(syn_50_data['epoch'], syn_50_data['metrics/mAP50(B)'], label='syn_50 mAP50', color='#6baed6')
plt.title('mAP50 Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('mAP50')
plt.legend()
plt.grid(True)
plt.show()
sns.reset_orig()
```
